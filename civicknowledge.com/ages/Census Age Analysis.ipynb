```python
from ambry import get_library
l = get_library()
cm = l.partition('census.gov-acs_geofile-schemas-2009e-column_meta-2014-5')
tm = l.partition('census.gov-acs_geofile-schemas-2009e-table_meta-2014-5')
```

```python
age_tables = []
for row in tm:
    if '"age"' in row.topics:
        age_tables.append(row.table_id)
```

```python
age_cols = []
for row in cm:
    if row.table_id in age_tables:
        age_cols.append(row.column_title)
```

```python
age_cols = sorted(set(age_cols))
```

```python
age_cols = filter(lambda r: 'years' in r, age_cols)
```
